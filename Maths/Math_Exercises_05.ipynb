# Exercise 3

Let f(x) = x<sup>4</sup> + 5x<sup>3</sup> − 7x  
Use the gradient descent method to find a (local) minimum x<sub>min</sub> for
the function f. Calculate the derivative f'(x) by hand, and find x<sub>min</sub>iteratively using the formula  
x<sub>i+1</sub> = x<sub>i</sub> − αf'(x<sub>i</sub>).  
You can do this by hand using a calculator, or write a program to
do calculations using a computer.


```python
# To get a more accurate answer, add more points to linspace
x = np.linspace(-5, 2, 1000)
y = x**4 + 5 * x**3 - 7 * x

plt.grid()
plt.plot(x,y)
plt.show()
```

<code>isMinimum</code> "algorithm" found from [stackoverflow](https://stackoverflow.com/questions/4624970/finding-local-maxima-minima-with-numpy-in-a-1d-numpy-array).

```python
isMinimum = np.r_[True, y[1:] < y[:-1]] & np.r_[y[:-1] < y[1:], True]

arr = []
for index, min in enumerate(isMinimum):
    if (min == True): arr.append(x[index])

print ("Local minima:", arr)
```

```python
plt.grid()
plt.plot(x,y)
plt.xticks(arr)
plt.show()
```

X ticks are the local minima.
